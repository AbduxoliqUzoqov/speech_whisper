```python
import torch
import numpy as np
from transformers import WhisperForConditionalGeneration, WhisperProcessor

# 1. Model va protsessorni yuklash
model_path = "/content/whisper-uz-final"
model = WhisperForConditionalGeneration.from_pretrained(model_path).to("cuda" if torch.cuda.is_available() else "cpu")
processor = WhisperProcessor.from_pretrained(model_path)

# 2. Test datasetdan bitta namuna olish
sample = next(iter(test_ds))

# 3. input_features-ni tensorga o'tkazish
# Listni numpy arrayga, keyin torch tensorga aylantiramiz
input_features = torch.tensor([sample["input_features"]]).to(model.device)

# 4. Matn generatsiya qilish
with torch.no_grad():
    predicted_ids = model.generate(input_features, language="uzbek", task="transcribe")

# 5. Natijani dekodlash
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]

# 6. Solishtirish
text_key = "text" if "text" in sample else "sentence"
print("-" * 30)
print(f"Haqiqiy matn: {sample[text_key]}")
print(f"Model natijasi: {transcription}")
print("-" * 30)
```

```python
import torch
import librosa
import numpy as np
from IPython.display import HTML, display
from google.colab.output import eval_js
from base64 import b64decode
import scipy.io.wavfile as wavfile
import ffmpeg

# 1. Ovoz yozish va faylga saqlash funksiyasi
def record_and_save_audio(filename="my_voice_record.wav"):
    RECORD_JS = """
    const sleep = time => new Promise(resolve => setTimeout(resolve, time));
    const b2text = b => new Promise(resolve => {
        const reader = new FileReader();
        reader.onloadend = () => resolve(reader.result);
        reader.readAsDataURL(b);
    });
    var recorder;
    var data;
    async def record(seconds) {
        const stream = await navigator.mediaDevices.getUserMedia({audio:true});
        recorder = new MediaRecorder(stream);
        const chunks = [];
        recorder.ondataavailable = e => chunks.push(e.data);
        recorder.start();
        await sleep(seconds * 1000);
        recorder.stop();
        await sleep(200);
        const blob = new Blob(chunks);
        data = await b2text(blob);
        return data;
    }
    """
    display(HTML(f"<script>{RECORD_JS.replace('async def', 'async function')}</script>"))
    print(f"10 soniya davomida gapiring...")
    audio_data = eval_js("record(7)")

    # Dekodlash va saqlash
    binary = b64decode(audio_data.split(',')[1])

    # Ffmpeg orqali WAV formatiga o'tkazish
    process = (ffmpeg
        .input('pipe:0')
        .output(filename, format='wav', acodec='pcm_s16le', ac=1, ar='16k')
        .overwrite_output()
        .run_async(pipe_stdin=True, quiet=True)
    )
    process.communicate(input=binary)
    print(f"Ovoz saqlandi: {filename}")

# 2. Ovozni yozamiz

```

```python
record_and_save_audio("my_voice_record.wav")

# 3. Saqlangan faylni model orqali tekshiramiz
print("Model tahlil qilmoqda...")
speech, sr = librosa.load("my_voice_record.wav", sr=16000)
inputs = processor(speech, sampling_rate=16000, return_tensors="pt").input_features.to(model.device)

with torch.no_grad():
    generated_ids = model.generate(inputs, language="uzbek", task="transcribe")

transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("-" * 30)
print(f"Saqlangan fayldagi natija: {transcription}")
print("-" * 30)
```
